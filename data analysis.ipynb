```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
stocks = pd.read_csv("stock data.csv")
stocks = stocks.drop(columns=['PERMNO','RET']).dropna()
stocks = stocks[stocks['TICKER']!='LEN']

prices = stocks.drop_duplicates().pivot(index='date', columns='TICKER', values='PRC')
prices.index = pd.to_datetime(prices.index, format='%m/%d/%Y')
prices = prices.sort_index()
prices.reset_index(inplace=True)
prices = prices.dropna(axis=1)

indexes = pd.read_csv("index data.csv")
indexes['date'] = pd.to_datetime(indexes['date'])
indexes = indexes.drop(columns=['PERMNO','RET']).dropna().pivot(index='date', columns='TICKER', values='PRC')
indexes.reset_index(inplace=True)
```

```python
stockToIndexMap = {'CMCSA':'XLY', 'SBUX':'XLY', 'F':'XLY', 'D':'XLU', 
    'AEP':'XLU', 'PG':'XLP', 'COKE':'XLP','PEP':'XLP', 'CL':'XLP','MSFT':'XLK',
    'INTC':'XLK', 'CSCO': 'XLK', 'ORCL':'XLK', 'T':'XLK', 'VZ':'XLK', 
    'HON':'XLI', 'JPM':'XLF', 'WFC':'XLF', 'IP':'XLB', 'WY':'IYR',
    'PLD':'IYR', 'VTR':'IYR', 'NTRS':'KBE', 'FITB':'KBE', 'C':'KBE',
    'PNC':'KBE', 'STT':'KCE', 'IVZ':'KCE', 'LAZ':'KCE', 'TOL':'XHB',
    'JCI':'XHB', 'MAS':'XHB', 'PHM':'XHB', 'ODP':'XRT',
    'URBN':'XRT', 'ADI':'XSD', 'HAL':'XES', 'PTEN':'XES'}
```

```python
for stock in stockToIndexMap.keys():
    data = (prices.copy()[['date',stock]].merge(indexes.copy()[['date',stockToIndexMap[stock]]], on='date', how='inner'))
    data[stock] = data[stock].abs()
    data[stockToIndexMap[stock]] = data[stockToIndexMap[stock]].abs()
    data.to_csv(f'stockandETF/{stock}.csv', index=False)
```
